```python

from nltk.stem import PorterStemmer
nltk.download('punkt')
porter = PorterStemmer()
l_words = ['dogs', 'programming', 'programs', 'programmed', 'cakes', 'indices', 'matrices']
for word in l_words:
    print(f'{word} \t -> {porter.stem(word)}'.expandtabs(15))
```
